```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.utils import Bunch


iris: Bunch = load_iris()  # type: ignore
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)
sns.pairplot(iris_df, hue="species")
plt.suptitle("Pair Plot of the Iris Dataset", size=20, y=1.02)
plt.show()
```

```python
corr = pd.DataFrame(data=iris.data, columns=iris.feature_names).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Iris Features")
plt.show()
```

```python
pca_result = PCA(n_components=4).fit_transform(iris.data)
```

```python
result_df = pd.DataFrame(data=pca_result)
result_df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)
sns.pairplot(result_df, hue="species")
plt.suptitle("Pair Plot of the Iris Dataset", size=20, y=1.02)
plt.show()
```

```python
corr = pd.DataFrame(data=pca_result).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Iris Features")
plt.show()
```
